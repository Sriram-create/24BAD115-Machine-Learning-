```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Download dataset
path = kagglehub.dataset_download("carrie1/ecommerce-data")

# Load dataset
df = pd.read_csv(path + "/data.csv", encoding="ISO-8859-1")
```

```python
# Dataset inspection
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Missing values
df.isnull().sum()
```

```python
# Sales per product
product_sales = df.groupby("Description")["Quantity"].sum().head(10)

# Bar chart
product_sales.plot(kind="bar")
plt.title("Top Products by Quantity Sold")
plt.xlabel("Product")
plt.ylabel("Quantity")
plt.show()
```

```python
# Line chart
product_sales.plot(kind="line")
plt.title("Sales Trend of Top Products")
plt.xlabel("Product")
plt.ylabel("Quantity")
plt.show()
```
